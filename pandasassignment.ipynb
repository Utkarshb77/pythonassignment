PART 2: World Population Data

```python
df = pd.read_csv('world_population.csv')
```

```python
df.head(10)
```

```python
print(df.isnull().sum())
df.dropna(inplace=True)
```

```python
top5_pop = df.sort_values(by='World Population Percentage', ascending=False).head(5)
print(top5_pop[['World Population Percentage']])
```

```python
filtered = df[(df['2022 Population'] > 100000000) & (df['Area (kmÂ²)'] < 500000)]
print(filtered[['Country/Territory', '2022 Population', 'Area (kmÂ²)']])
```

```python
df['Density (per kmÂ²)'] = df['2022 Population'] / df['Area (kmÂ²)']
sorted_density = df.sort_values(by='Density (per kmÂ²)', ascending=False)
print(sorted_density[['Country/Territory', 'Density (per kmÂ²)']])
```

PART 4: Students Performance

```python
stp = pd.read_csv('studentsPerformance.csv')
stp.head(1)
```

```python
print(stp['math score'].mean())
print(stp['reading score'].mean())
print(stp['writing score'].mean())
```

```python
writing_avg_by_gender = stp.groupby('gender')['writing score'].mean()
better_gender = writing_avg_by_gender.idxmax()
better_gender.capitalize()
```

```python
stp['total_score'] = stp['math score'] + stp['reading score'] + stp['writing score']
stp['percentage'] = stp['total_score'] / 3
stp.head(1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avg_scores_by_parent = stp.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
avg_scores_by_parent.plot(kind='bar', figsize=(10, 5))
plt.title('Average Scores by Parental Level of Education')
plt.ylabel('Average Score')
plt.show()
```

```python
highscorers = stp[(stp['math score'] > 90) & (stp['reading score'] > 90 )&( stp['writing score'] > 90)]
highscorers
```

PART 5: Flight Price pridiction Dataset

```python
clean = pd.read_csv('Clean_Dataset.csv')
```

```python
clean.head(2)
```

```python
clean.info()
```

```python
clean.describe()
```

```python
from datetime import datetime
clean['Journey_Date'] = pd.to_datetime(datetime.today()) + pd.to_timedelta(clean['days_left'], unit='D')
clean['Journey_Day'] = clean['Journey_Date'].dt.day
clean['Journey_Month'] = clean['Journey_Date'].dt.month
clean.head(2)
```

```python
most_common_airline = clean['airline'].value_counts().idxmax()
most_common_source = clean['source_city'].value_counts().idxmax()
print(most_common_airline)
most_common_source
```
